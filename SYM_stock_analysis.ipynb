```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

ticker = 'SYM'
sym_data = yf.download(ticker, period="max")
print("")
print(sym_data.head())

plt.figure(figsize=(14, 7))
plt.plot(sym_data.index, sym_data['Close'], label='SYM Close Price')
plt.title('Symbotic Inc. (SYM) Historical Stock Price', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend()
plt.grid(True) 
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import pprint

url = 'https://en.wikipedia.org/wiki/Symbotic'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

infobox = soup.select_one('table.infobox')

company_data = {}

if infobox:
    rows = infobox.find_all('tr')
    for row in rows:
        header = row.find('th')
        data = row.find('td')
        if header and data:
            key = header.text.strip()
            value = data.text.strip()
            company_data[key] = value
            
    description_parts = []
    p_count = 0
    for p_tag in infobox.find_next_siblings('p'):
        if p_count < 2:
            description_parts.append(p_tag.text.strip())
            p_count = p_count + 1
        else:
            break
            
    if description_parts:
        company_data['Description'] = ''.join(description_parts)

if company_data:
    print('---  ---')
    pprint.pprint(company_data)
else:
    print(' infobox ')
```

```python
company_type = company_data.get('Company type', '')
industry = company_data.get('Industry', '')
headquarters = company_data.get('Headquarters', '')
key_people = company_data.get('Key people', '')
description = company_data.get('Description', '')

summary_part1 = f"""
: Symbotic Inc.

:
- : {company_type}
- : {industry}
- : {headquarters}
- : {key_people}
"""

description_title = ':'

import re

#  ".[1]" 
description = re.sub(r'\.\[\d+\]', '.', description)
description_body = description

full_report = summary_part1.strip() + '\n\n' + description_title + '\n' + description_body

print(full_report)
```

```python
import yfinance as yf

sym_ticker = yf.Ticker('SYM')

income_statement = sym_ticker.financials

balance_sheet = sym_ticker.balance_sheet

print('---  (Income Statement) ---')
print(income_statement)

print('')
print('---  (Balance Sheet) ---')
print(balance_sheet)
```

```python
income_data_to_plot = income_statement.loc[['Total Revenue', 'Net Income']].transpose()
income_data_to_plot = income_data_to_plot.drop(index='2020-09-30 00:00:00')
income_data_to_plot = income_data_to_plot.iloc[::-1]

income_data_to_plot.plot(kind='bar', figsize=(12, 7))

plt.title('Symbotic Annual Revenue vs. Net Income')
plt.xlabel('Year')
plt.ylabel('Amount (in USD)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
balance_data_to_plot = balance_sheet.loc[['Total Assets', 'Total Liabilities Net Minority Interest']].transpose()
balance_data_to_plot = balance_data_to_plot.drop(index='2020-09-30 00:00:00')
balance_data_to_plot = balance_data_to_plot.iloc[::-1]

balance_data_to_plot.plot(kind='bar', figsize=(12, 7))

plt.title('Symbotic Total Assets vs. Total Liabilities')
plt.xlabel('Year')
plt.ylabel('Amount (in USD)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
